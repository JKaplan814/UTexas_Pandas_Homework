### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Find number of unique players
players_count = purchase_data["SN"].nunique()

#Create data frame
total_players = pd.DataFrame({"Total Players":[players_count]})

#Print data frame
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Create values for the data frame
unique_items = purchase_data["Item Name"].nunique()
average_price = purchase_data["Price"].mean()
num_purchases = purchase_data['Purchase ID'].count()
total_revenue = purchase_data['Price'].sum()

#Create the data frame
summary_frame = pd.DataFrame({
    "Number of Unique Items":[unique_items],'Average Price':[average_price],'Number of Purchases':[num_purchases],'Total Revenue':[total_revenue]
})

#Edit dollar values to appear as dollar values
summary_frame['Average Price'] = summary_frame['Average Price'].astype(float).map("${:,.2f}".format)
summary_frame['Total Revenue'] = summary_frame['Total Revenue'].astype(float).map("${:,.2f}".format)

#Print the frame
summary_frame
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Create dataframe to remove duplicate usernames so they aren't counted as players twice
unique_players = purchase_data.drop_duplicates('SN')

#Determine count of player's gender
genders = unique_players['Gender'].value_counts()
males = genders[0]
females = genders[1]
other = genders[2]

#Calculate percentages
male_pct = (males / players_count) * 100
female_pct = (females / players_count) * 100
other_pct = (other / players_count) * 100

#Create dataframe
gender_summary = pd.DataFrame({
    'Percent of Players':[male_pct,female_pct,other_pct],
    'Total Count':[males,females,other]
})

#Replace index and format percentages
gender_summary = gender_summary.rename(index = {0:'Male',1:'Female',2:'Other/Non-Disclosed'})
gender_summary['Percent of Players'] = gender_summary['Percent of Players'].astype(float).map('{:.2f}%'.format)
gender_summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Determine purchase counts
genders_purchases = purchase_data['Gender'].value_counts()
male_purchases = genders_purchases[0]
female_purchases = genders_purchases[1]
other_purchases = genders_purchases[2]

#Create male df and analyze
male_data = purchase_data.loc[purchase_data['Gender'] == 'Male',:]
male_total = male_data['Price'].sum()
male_average_price = male_total / male_purchases
male_average_purchase = male_total / males

#Create female df and analyze
female_data = purchase_data.loc[purchase_data['Gender'] == 'Female',:]
female_total = female_data['Price'].sum()
female_average_price = female_total / female_purchases
female_average_purchase = female_total / females

#Create other df and analyze
other_data = purchase_data.loc[purchase_data['Gender'] == 'Other / Non-Disclosed',:]
other_total = other_data['Price'].sum()
other_average_price = other_total / other_purchases
other_average_purchase = other_total / other

#create df
gender_purchase_summary = pd.DataFrame({
    'Purchase Count':[male_purchases,female_purchases,other_purchases],
    'Average Purchase Price':[male_average_price,female_average_price,other_average_price],
    'Total Purchase Value':[male_total,female_total,other_total],
    'Avg Purchase Total per Person':[male_average_purchase,female_average_purchase,other_average_purchase]   
})

#change index and format numbers
gender_purchase_summary = gender_purchase_summary.rename(index = {0:'Male',1:'Female',2:'Other/Non-Disclosed'})
gender_purchase_summary.index.name = 'Gender'
gender_purchase_summary['Average Purchase Price'] = gender_purchase_summary['Average Purchase Price'].astype(float).map('${:.2f}'.format)
gender_purchase_summary['Total Purchase Value'] = gender_purchase_summary['Total Purchase Value'].astype(float).map('${:.2f}'.format)
gender_purchase_summary['Avg Purchase Total per Person'] = gender_purchase_summary['Avg Purchase Total per Person'].astype(float).map('${:.2f}'.format)

gender_purchase_summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Place unique players into age bins
#Solution used the purchase data here instead of unique players. This will skew the percentage of players to inaccuracy because some players are counted multiple times.
unique_players['Age Group'] = pd.cut(unique_players['Age'],age_bins,labels=group_names)

#Group by age bin and create dataframe columns
age_groups = unique_players.groupby('Age Group')
age_group_count = age_groups['Age'].count().rename('Total Count')
age_group_percentage = (age_group_count / players_count) * 100
age_group_percentage = age_group_percentage.rename('Percent of Players')

#Create dataframe
age_group_summary = pd.concat([age_group_percentage,age_group_count],axis=1)
age_group_summary['Percent of Players']=age_group_summary['Percent of Players'].map('{:.2f}%'.format)
age_group_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Place purchase data into age bins
purchase_data['Age Group'] = pd.cut(purchase_data['Age'],age_bins,labels=group_names)
purchase_age_groups = purchase_data.groupby('Age Group')

#Calculate dataframe values
pag_count = purchase_age_groups['Price'].count().rename('Purchase Count')
pag_avg_price = purchase_age_groups['Price'].mean().rename('Average Purchase Price').map('${:.2f}'.format)
pag_total_value = purchase_age_groups['Price'].sum().rename('Total Purchase Value')
pag_total_per_person = pag_total_value / age_group_count
pag_total_per_person = pag_total_per_person.rename('Average Purchase Total per Person').map('${:.2f}'.format)
pag_total_value = pag_total_value.map('${:.2f}'.format)

#Create dataframe
pag_summary = pd.concat([pag_count, pag_avg_price, pag_total_value, pag_total_per_person],axis = 1)
pag_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by SN
purchases = purchase_data.groupby('SN')

#Create dataframe values
purchase_count = purchases['SN'].count().rename('Purchase Count')
avg_purchase_price = purchases['Price'].mean().rename('Average Purchase Price').map('${:.2f}'.format)
total_purchase_value = purchases['Price'].sum().rename('Total Purchase Value').map('${:.2f}'.format)

#Create dataframe
sn_purchase_summary = pd.concat([purchase_count, avg_purchase_price, total_purchase_value],axis=1)
sn_purchase_summary = sn_purchase_summary.sort_values(by='Purchase Count',ascending=False)
sn_purchase_summary.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create item dataframe
items_loc = purchase_data.loc[:,['Item ID','Item Name','Price']]
items = items_loc.groupby(['Item ID','Item Name'])

#Create dataframe values
items_count = items['Price'].count().rename('Purchase Count')
item_price = items['Price'].mean().rename('Item Price').map('${:.2f}'.format)
item_total_value = items['Price'].sum().rename('Total Purchase Value').map('${:.2f}'.format)

#Create dataframe
item_popularity_summary = pd.concat([items_count, item_price, item_total_value],axis=1)
item_popularity_summary = item_popularity_summary.sort_values(by='Purchase Count',ascending=False)

item_popularity_summary.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Sort above table
#Total Purchase Value wasn't sorting properly even though the other two columns were. Removed string values and converted back to float and it sorted properly.
item_popularity_summary['Total Purchase Value'] = item_popularity_summary['Total Purchase Value'].str.replace('[/$,]','').astype(float)

item_profitability_summary = item_popularity_summary.sort_values(by='Total Purchase Value',ascending=False)
item_profitability_summary['Total Purchase Value']=item_profitability_summary['Total Purchase Value'].map('${:.2f}'.format)

item_profitability_summary.head()
```
